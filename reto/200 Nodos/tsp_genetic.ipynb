```python
import numpy as np, random, operator
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dist_matrix0.csv', index_col=0)
```

```python
df
num_cities = len(df)
```

```python
def fitness(chromosome,df):
    fitness = 0
    for i in range(len(chromosome)):
        fitness += df.iloc[chromosome[i-1],chromosome[i]]
    return -fitness
```

```python
class Agent():
    def __init__(self, num_cities):
        self.create_route()
        self.fitness = fitness(self.route,df)
    
    def create_route(self):
        self.route = list(random.sample(range(len(df)), len(df)))
    def mutate(self):
        for swaping in range(len(self.route)):
            if(random.random() < 0.1):
                swap_with = int(random.random() * len(self.route))
                city1 = self.route[swaping]
                city2 = self.route[swap_with]
                self.route[swaping] = city2
                self.route[swap_with] = city1
        self.fitness = fitness(self.route,df)
        
```

```python
#Se toma la mitad de la poblaci칩n con mejor fitness
def selection(population):
    #Proceso de seleccion generico, no creo que sea el m치s adecuado
    sorted_population = sorted(population, key=lambda x: x.fitness, reverse=True)
    return sorted_population[:len(population)//2]
    
#Se realiza el cruce de los cromosomas de los agentes
def crossover(population):
    #Funci칩n usada en PROBLEMA ANTERIOR, son parece que sea la mejor manera de hacer crossover
    offspring = []
    
    # Convert the elements list to a numpy array
    arpopulation = np.array(population)
    
    # Shuffle the elements using numpy's shuffle
    np.random.shuffle(arpopulation)

    # Check if the number of elements is even
    if len(arpopulation) % 2 != 0:
        raise ValueError("The number of elements must be even to form pairs.")
    
    # Reshape the array into pairs
    arparents = arpopulation.reshape(-1, 2)
    
    # Convert the numpy array of pairs back to a list of lists
    parents = arparents.tolist()
    
    for parent in parents:
        parent1 = parent[0] 
        parent2 = parent[1]
        child1 = Agent(num_cities)
        child2 = Agent(num_cities)
        geneA = int(random.random() * len(parent1.route))
        geneB = int(random.random() * len(parent1.route))
        startGene = min(geneA, geneB)
        endGene = max(geneA, geneB)
        for i in range(startGene, endGene):
            child1.route[i] = parent1.route[i]
            child2.route[i] = parent2.route[i]
        for i in range(len(parent2.route)):
            if parent2.route[i] not in child1.route:
                for j in range(len(parent2.route)):
                    if child1.route[j] == -1:
                        child1.route[j] = parent2.route[i]
                        break
            if parent1.route[i] not in child2.route:
                for j in range(len(parent1.route)):
                    if child2.route[j] == -1:
                        child2.route[j] = parent1.route[i]
                        break
        childs = [child1, child2]
        mutated_childs = mutation(childs)
        # Check the fitness of the mutated children
        for child in mutated_childs:
            child.fitness = fitness(child.route,df)
        mutated_childs = sorted(mutated_childs, key=lambda x: x.fitness)
        mutated_childs = mutated_childs
        offspring.extend(mutated_childs)
    # print(len(offspring))

        
        
    return offspring
# Se realiza la probabilidad de mutaci칩n
def mutation(population):
    for entity in population:
        if random.random() < 0.5: #Probabilidad de mutacion
            entity.mutate()
    return population
```

```python
#population_size = int(0.01 * len(weight)**20)
from numpy import cross

# Se inicializa la poblaci칩n
population_size = 100
population = [Agent(num_cities) for i in range(population_size)]
counter = 0
past_agent = 0
epoch = 0
#N칰mero de epocas

while epoch < 1000:
    epoch += 1
    babies = crossover(population)
    population.extend(babies)
    population = selection(population)
    #Look for the best agent
    best_agent = max(population, key=lambda x: x.fitness)
    if best_agent.fitness == past_agent:
        counter += 1
    else:
        counter = 0
    past_agent = best_agent.fitness
    print("Epoca:"+ str(epoch))
    print("Este es el fitness mejor agente: " + str(best_agent.fitness))
    #Save the best agent of all time
    if epoch == 1:
        best_agent_ever1 = best_agent
    else:
        if best_agent.fitness > best_agent_ever1.fitness:
            best_agent_ever1 = best_agent
```
